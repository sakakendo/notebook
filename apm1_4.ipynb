```
#[演習3] (2)
import numpy as np
import matplotlib.pyplot as plt
import matplotlib.font_manager

plt.figure( figsize=(15,6))
x= np.linspace(-4,4,100)

y=1
b=lambda n:(4*(1-np.cos(np.pi*n))/np.power(np.pi*n,2))

for n in range(1,6):
  y+=b(n)*np.cos(n*np.pi/2*x)
plt.plot(x,y)
plt.xlabel("t[s]")
plt.ylabel("f(t)")
plt.xticks([-4,-2,0,2,4])

plt.show()

```
