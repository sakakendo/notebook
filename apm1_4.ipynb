# [演習2]  



```
#@title (2)
#fig8.10

import numpy as np
import matplotlib.pylab as plt
#fig = plt.figure(figsize=(3,1),dpi=200)

x=np.linspace(-np.pi*2,2*np.pi,300)
f=np.pi/2
plt.xticks([-2*np.pi,-1*np.pi,0,np.pi,6.28],['-2π','-π','0','π','2π'])
plt.yticks(range(-2,4),range(-2,4))
a=lambda n:(np.sin(3/4*np.pi*n)+np.sin(1/4*np.pi*n))/n
b=lambda n:-(np.cos(3/4*np.pi*n)-np.cos(1/4*np.pi*n))/n
for n in range(1,6):
  f+=a(n)*np.cos(n*x)+b(n)*np.sin(n*x)
plt.plot(x,f)

plt.xlabel('angle[rad]')
plt.ylabel('f(x)')
plt.show()
```

```
#@title (1)
import numpy as np
import matplotlib.pyplot as plt
import matplotlib.font_manager

plt.figure( figsize=(15,6))
x= np.linspace(-2*np.pi,2*np.pi,100)

y=1
b=lambda n:(4*(1-np.cos(np.pi*n))/np.power(np.pi*n,2))

for n in range(1,6):
  y+=b(n)*np.cos(n*np.pi/2*x)
plt.plot(x,y)
plt.xlabel("t[s]")
plt.ylabel("f(t)")
plt.xticks([-2*np.pi,-1*np.pi,0,np.pi,6.28],['-2π','-π','0','π','2π'])
plt.show()

```
