```
#@title [例題10-6]の図 { vertical-output: true, display-mode: "form" }
#[例題10-6]の図
import matplotlib.pylab as plt
import numpy as np
plt.yticks([0,np.pi],[0,'π'])

x=np.linspace(-5*np.pi,5*np.pi,10000)


y=2/(4+np.power(x,2))
plt.subplot(221)
plt.title("Re{F(ω)}")
#print([0.1*n for n in range(10)])
#print([np.pi*n for n in range(5)])
#plt.xticks([np.pi*n for n in range(5)],[str(np.pi*n) for n in range(5)])
plt.xticks([-5*np.pi,0,5*np.pi],['-5π','0','5π'])
plt.plot(x,y)

y=-x/(4+np.power(x,2))
plt.subplot(222)
plt.title("Im{F(ω)}")
plt.xticks([-5*np.pi,0,5*np.pi],['-5π','0','5π'])
plt.plot(x,y)

y=1/np.sqrt((4+np.power(x,2)))
plt.subplot(223)
plt.title("|F(ω)|")
plt.xticks([-5*np.pi,0,5*np.pi],['-5π','0','5π'])
plt.plot(x,y)

y=np.arctan(-x/2)
plt.subplot(224)
plt.title("θ")
plt.xticks([-5*np.pi,0,5*np.pi],['-5π','0','5π'])
plt.plot(x,y)

plt.subplots_adjust(wspace=0.4, hspace=0.6)
plt.show()
```

```
#@title [例題10-7]の図
#[例題10-7]の図

import matplotlib.pylab as plt
import numpy as np

x=np.linspace(-5*np.pi,5*np.pi,1000)

y=4*(1/(4+np.power(x,2)))
plt.subplot(221)
plt.title("Re{F(ω)}")
plt.xticks([-5*np.pi,0,5*np.pi],['-5π','0','5π'])
plt.plot(x,y)

plt.subplot(223)
plt.title("|F(ω)|")
plt.xticks([-5*np.pi,0,5*np.pi],['-5π','0','5π'])
plt.plot(x,y)


plt.subplot(222)
plt.title("Im{F(ω)}")
plt.xticks([-5*np.pi,0,5*np.pi],['-5π','0','5π'])
plt.plot(x,x*0)

plt.subplot(224)
plt.title("θ(ω)")
plt.xticks([-5*np.pi,0,5*np.pi],['-5π','0','5π'])
#plt.yticks([np.pi,0,np.pi],['-π','0','π'])
plt.plot(x,x*0)

plt.subplots_adjust(wspace=0.4, hspace=0.6)
plt.show()
```

```
#@title [例題10-8]の図
# [例題10-8]の図
import matplotlib.pylab as plt
import numpy as np

x=np.linspace(-5*np.pi,5*np.pi,1000)

#Re{F(w)}
y=(16+2*np.power(x,2))/(np.power(x,4)+64)
plt.subplot(221)
plt.title("Re{F(ω)}")
plt.xticks([-5*np.pi,0,5*np.pi],['-5π','0','5π'])
plt.plot(x,y)

#Im{F(w)}
y=-np.power(x,3)/(np.power(x,4)+64)
plt.subplot(222)
plt.title("Im{F(ω)}")
plt.xticks([-5*np.pi,0,5*np.pi],['-5π','0','5π'])
plt.plot(x,y)

#|F(w)|
y=np.sqrt((np.power(x,2)+4)/(np.power(x,4)+64))
plt.subplot(223)
plt.title("|F(ω)|")
plt.xticks([-5*np.pi,0,5*np.pi],['-5π','0','5π'])
plt.plot(x,y)

y=-np.arctan(np.power(x,3)/(16+2*np.power(x,2)))
plt.subplot(224)
plt.title("θ(ω)")
plt.xticks([-5*np.pi,0,5*np.pi],['-5π','0','5π'])
plt.plot(x,y)

plt.subplots_adjust(wspace=0.4, hspace=0.6)
plt.show()
```
