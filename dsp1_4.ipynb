```
#plot file

import matplotlib.pylab as plt
import numpy as np

plt.figure( figsize=(32,8))

def plotWdata(fname):
  data = np.genfromtxt(fname, dtype = None, delimiter = "\n")
  print(fname)
  plt.plot(np.arange(len(data)),data,label=fname)  
def plotRdata(fname):
  data = np.genfromtxt(fname, dtype = None, delimiter = "\n")
  print(fname)
  plt.plot(np.arange(len(data)),data*30,label=fname)  
  
plotWdata('wdata1.txt')
plotWdata('wdata2.txt')
plotRdata('r11.txt')
plotRdata('r12.txt')
plt.legend(loc='upper left',prop={'size':14})
plt.show()

```
