```
#@title [演習2]
import math
import numpy as np
from matplotlib import pyplot

x = np.linspace(0, 1000, 100000)  #0から2πまでの範囲を100分割したnumpy配列
pyplot.figure(figsize=(16/2,9/2),dpi=100)
#pyplot.plot(x, np.sin(1/50*np.pi*x))   #10Hz
pyplot.plot(x, np.sin(18*np.pi*x/1000)) #9Hz
pyplot.plot(x, -np.sin(2*np.pi*x/1000)) #1Hz
pyplot.xticks(range(0,1001,100),range(0,10001,100))

pyplot.xlabel('t[ms]')
pyplot.ylabel('y(t)')
pyplot.show()
```

```
#@title (2)
import math
import numpy as np
from matplotlib import pyplot
pyplot.figure(figsize=(10,1),dpi=100)
pyplot.plot([-200*np.pi,-100*np.pi,-100*np.pi,100*np.pi,100*np.pi,200*np.pi],[0,0,1,1,0,0]) #9Hz

pyplot.xticks([-100*np.pi,100*np.pi],["-100π","100π"])

pyplot.xlabel('ω[rad/s]')
pyplot.ylabel('H(ω)')
pyplot.show()
```

```
#@title (2)
import math
import numpy as np
from matplotlib import pyplot
pyplot.figure(figsize=(5,5),dpi=100)
pyplot.plot([-1,1],[1,-1]) #9Hz

pyplot.xticks([-1,0,1],[1,0,-1])
pyplot.yticks([-1,0,1],[1,0,-1])

pyplot.xlabel('ω[rad/s]')
pyplot.ylabel('θ(ω)')
pyplot.show()
```
