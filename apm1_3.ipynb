```
#@title デフォルトのタイトル テキスト
#(2) fig8.10
import numpy as np
import matplotlib.pylab as plt
fig = plt.figure(figsize=(5,3),dpi=200)

x=np.linspace(-np.pi*5/4,2*np.pi,300)
f=np.pi/2
n=1
a=lambda n:(np.sin(3/4*np.pi*n)+np.sin(1/4*np.pi*n))/n
b=lambda n:-(np.cos(3/4*np.pi*n)-np.cos(1/4*np.pi*n))/n
for n in range(1,10):
  f+=a(n)*np.cos(n*x)+b(n)*np.sin(n*x)
  if n in [1,2]:
    plt.plot(x,a(n)*np.cos(n*x))
    plt.plot(x,b(n)*np.sin(n*x))
  plt.plot(x,f)
  #if n in [1,2,3,10,30,90,999]:  plt.plot(x,f)

plt.xlabel('angle[rad]')
plt.ylabel('f(x)')
plt.show()
```
