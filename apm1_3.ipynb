&=&-\frac{1}{2T}\left[ \frac{1}{3\omega_0}\cdot cos3\omega_0t+\frac{1}{\omega}cos(-\omega_0t) \right]^\pi_{-\pi} \\

$(2)\\ \begin{eqnarray}
&\ & \langle cos2\omega_0t,cos3\omega_0t\rangle\\ 
&=&\int^{\frac{T}{2}}_{-\frac{T}{2}} \cos2\omega_0t\cdot\cos3\omega_0t dt \\ 
&=&\frac{1}{2}\int^{\frac{T}{2}}_{-\frac{T}{2}} \cos5\omega_0t+\cos\omega_0t dt\\
&=&\frac{1}{2} \left[\frac{1}{5\omega_0}\cos5\omega_0t+\frac{1}{\omega_0}\cos\omega_0t \right]^{\frac{T}{2}}_{-\frac{T}{2}} \\

$(3)\\\begin{eqnarray}

&=&\frac{1}{T}\int_{-\frac{T}{2}}^{\frac{T}{2}}\sin\omega_0t\cdot\cos3\omega_0t dt \\
&=&\frac{1}{2T}\int_{-\frac{T}{2}}^{\frac{T}{2}}\sin4\omega_0t-\sin2\omega_0t dt \\
&=&\frac{1}{2T} \left[ -\frac{1}{4\omega_0}\cos4\omega_0t+\frac{1}{2\omega_0t}\cos2\omega_0t \right]_{-\frac{T}{2}}^{\frac{T}{2}} \\
&=&\frac{1}{2T} \{\{ -\frac{1}{4\omega_0}\cos4\frac{2\pi}{T}\cdot\frac{T}{2}+\frac{1}{2\omega_0t}\cos2\frac{2\pi}{T}\cdot\frac{T}{2}\} -\{-\frac{1}{4\omega_0}\cos4\frac{2\pi}{T}\cdot\frac{T}{2}+\frac{1}{2\omega_0t}\cos2\frac{2\pi}{T}\cdot\frac{T}{2}\}\} \\

&=&0
\end{eqnarray}$

$(4)\\\begin{eqnarray}

&=&-\frac{1}{3\omega_0T}\left[\cos{3\omega_0t} \right]^{\frac{T}{2}}_{-\frac{2}{T}}\\

$(6)\\

&\ &<1,\cos{\omega_0t}>\\

&=&\frac{1}{T}\frac{T}{2\pi}\left[ \sin{\omega_0t}\right]^{\frac{T}{2}}_{-\frac{2}{T}}\\
&=&\frac{1}{2\pi}\{\sin{\frac{2\pi}{T}\frac{T}{2}}-\sin{-\frac{2\pi}{T}\frac{T}{2}} \}\\
&=&\frac{1}{2\pi}\{\sin{\pi}-\sin{\pi} \}\\
&=&0
\end{eqnarray}

```
#@title (2)fig8.10

import numpy as np
import matplotlib.pylab as plt
#fig = plt.figure(figsize=(3,1),dpi=200)

x=np.linspace(-np.pi*2,2*np.pi,300)
f=np.pi/2
plt.xticks([-2*np.pi,-1*np.pi,0,np.pi,6.28],['-2π','-π','0','π','2π'])
plt.yticks(range(-2,4),range(-2,4))
a=lambda n:(np.sin(3/4*np.pi*n)+np.sin(1/4*np.pi*n))/n
b=lambda n:-(np.cos(3/4*np.pi*n)-np.cos(1/4*np.pi*n))/n
for n in range(1,6):
  f+=a(n)*np.cos(n*x)+b(n)*np.sin(n*x)
plt.plot(x,f)

plt.xlabel('angle[rad]')
plt.ylabel('f(x)')
plt.show()
```
