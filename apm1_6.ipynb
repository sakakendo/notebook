```
#ex1-4
import numpy as np
import matplotlib.pylab as plot

x=np.linspace(-10,10,1000)

plt.figure()
plt.title("F(ω)")
y=4/x*np.sin(2*x)
plt.plot(x,y)


plt.figure()
plt.title("Im{F(ω)}")
y=x*0
plt.plot(x,y)

plt.figure()
plt.title("|F(ω)|")
y=np.absolute(4/x*np.sin(2*x))
plt.plot(x,y)

plt.figure()
plt.title("θ")
plt.yticks([0,np.pi],[0,'π'])
plt.plot(x,np.where(x*np.sin(2*x) > 0, 0, np.pi))

plt.show()
```

```
#ex1-5

import numpy as np
import matplotlib.pylab as plt

x=np.linspace(-5,5,100)
plt.figure()
plt.title("F(ω)")
y=4/x**2*np.sin(x)**2
plt.plot(x,y)

plt.figure()
plt.title("Re{F(ω)}")
y=4/x**2*np.sin(x)**2
plt.plot(x,y)

plt.figure()
plt.title("Im{F(ω)}")
y=x*0
plt.plot(x,y)

plt.figure()
plt.title("θ")
y=x*0
plt.plot(x,y)

plt.show()
```

```
# 演習2
import numpy as np
import matplotlib.pylab as plt

plt.xticks([-5*np.pi,-4*np.pi,-3*np.pi,-2*np.pi,-1*np.pi,0,np.pi,np.pi*2,np.pi*3,np.pi*4,np.pi*5],['-5π','-4π','-3π','-2π','-π','0','-π','-2π','-3π','-4π','-5π'])

x=np.linspace(-5*np.pi,5*np.pi,200)
plt.title("Re{F(ω)}")
plt.plot(x,4/x*np.sin(x))        #Re

plt.figure()
plt.title("Im{F(ω)}")
plt.plot(x,np.linspace(0,0,200)) #Im

plt.figure()
plt.title("|F(ω)|")
plt.plot(x,np.absolute(4/x*np.sin(x)))        #|F(w)|

plt.figure()
plt.title("θ")
plt.plot(x,np.where(x*np.sin(x) > 0, 0, np.pi)) #theta
plt.savefig("theta.png")

plt.show()
```
