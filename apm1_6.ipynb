```
#@title [例題10-4]
#ex1-4
import numpy as np
import matplotlib.pylab as plt

x=np.linspace(-10,10,1000)

plt.subplot(221)
plt.title("F(ω)")
y=4/x*np.sin(2*x)
plt.plot(x,y)

plt.subplot(222)
plt.title("Im{F(ω)}")
y=x*0
plt.plot(x,y)

#plt.figure()
plt.subplot(223)
plt.title("|F(ω)|")
y=np.absolute(4/x*np.sin(2*x))
plt.plot(x,y)

#plt.figure()
plt.subplot(224)
plt.title("θ")
plt.yticks([0,np.pi],[0,'π'])
plt.plot(x,np.where(x*np.sin(2*x) > 0, 0, np.pi))

plt.subplots_adjust(wspace=0.4, hspace=0.6)

plt.show()
```

$F(\omega)=\int^\infty_{-\infty}f(t)\cdot e^{-j\omega t}dt\\=\int^2_{-2}2\cdot e^{-j \omega t} dt\\
=-\frac{2}{j \omega}\left[e^{-j\omega t} \right]^2_{-2}\\
=-\frac{2}{j\omega}(e^{^j2\omega}-e^{j2\omega})\\
=\frac{4}{\omega}\sin2\omega$<br><br><br><br><br><br><br><br><br><br><br><br>

```
#@title [例題10-5]
#ex1-5

import numpy as np
import matplotlib.pylab as plt

x=np.linspace(-5,5,100)
#plt.figure()
plt.subplot(221)
plt.title("F(ω)")
y=4/x**2*np.sin(x)**2
plt.plot(x,y)

plt.subplot(222)
plt.title("Re{F(ω)}")
y=4/x**2*np.sin(x)**2
plt.plot(x,y)

plt.subplot(223)
plt.title("Im{F(ω)}")
y=x*0
plt.plot(x,y)

plt.subplot(224)
plt.title("θ")
y=x*0
plt.plot(x,y)
plt.subplots_adjust(wspace=0.4, hspace=0.6)

plt.show()
```

$F(\omega) = \int_{-\infty}^{\infty} f(t)\cdot e^{-j\omega t}dt \\$
$=\int_{-2}^{0} (2+t)\cdot e^{-j\omega t}dt+\int_{0}^{2} (2-t)\cdot e^{-j\omega t}dt \\
= \int_{-2}^{0} 2\cdot e^{-j\omega t}dt+\int_{-2}^{0} t\cdot e^{-j\omega t}dt+\int_{0}^{2} 2\cdot e^{-j\omega t}dt-\int_{0}^{2} t\cdot e^{-j\omega t}dt \\
=-\frac{2}{j\omega}\left[e^{-j\omega t}\right]_{-2}^{2}+\{-\frac{2}{j\omega}\left[t\cdot e^{-j\omega t}\right]_{-2}^{0}+\frac{1}{j\omega}\int_{-2}^{0} e^{^j\omega t}dt\}-\{-\frac{2}{j\omega}\left[t\cdot e^{-j\omega t}\right]_{0}^{2}+\frac{1}{j\omega}\int_{0}^{2} e^{^j\omega t}dt\} \\
%=-\frac{2}{j\omega}(e^{-j2\omega}-e^{j2\omega}) \\
= -\frac{2}{j\omega}+\frac{2}{j\omega}e^{j2\omega}-\frac{2}{j\omega}e^{j2\omega}+\frac{1}{\omega^2}-\frac{1}{\omega^2}e^{j2\omega}-\frac{2}{j\omega}e^{-j2\omega}+\frac{2}{j\omega}+\frac{2}{j\omega}e^{-j2\omega}+\frac{1}{\omega^2}-\frac{1}{\omega^2}e^{j2\omega} \\
=\frac{1}{\omega^2}(2-e^{j2\omega}-e^{-j2\omega}) \\
= \frac{2}{\omega^2}(1-\cos{2\omega}) \\
=\frac{4}{\omega^2} \sin^2{\omega}$

```
#@title [演習2]
# 演習2
import numpy as np
import matplotlib.pylab as plt

plt.xticks([-5*np.pi,-4*np.pi,-3*np.pi,-2*np.pi,-1*np.pi,0,np.pi,np.pi*2,np.pi*3,np.pi*4,np.pi*5],['-5π','-4π','-3π','-2π','-π','0','-π','-2π','-3π','-4π','-5π'])

x=np.linspace(-5*np.pi,5*np.pi,200)

plt.subplot(221)
plt.title("Re{F(ω)}")
plt.plot(x,4/x*np.sin(x))        #Re

plt.subplot(222)
plt.title("Im{F(ω)}")
plt.plot(x,np.linspace(0,0,200)) #Im

plt.subplot(223)
plt.title("|F(ω)|")
plt.plot(x,np.absolute(4/x*np.sin(x)))        #|F(w)|

plt.subplot(224)
plt.title("θ")
plt.plot(x,np.where(x*np.sin(x) > 0, 0, np.pi)) #theta
plt.savefig("theta.png")

plt.subplots_adjust(wspace=0.4, hspace=0.6)

plt.show()
```
